#  Ontario House price - Between 2019 to 2020

```python
#Importing Libraries
import numpy as np
import pandas as pd
import hvplot.pandas
%matplotlib inline
from sklearn import metrics
import seaborn as sns
import os
import matplotlib.pyplot as plt
import plotly.express as px
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read csv contains house prices in GTA
df = pd.read_csv("Data/House_price_GTA.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)

# Use GTA Composite Data
df_gta = df[["Composite_Benchmark_SA_GTA"]]

# Use GTA Composite Data - from 2019 to 2020
df_gta_19_20 = df_gta["2019-01-01":"2020-09-01"]

df_gta_19_20.head()
```

```python
# GTA House price from 2019 to 2020 plot
gta_graph = df_gta_19_20.hvplot(title = "GTA House price in 2019 to 2020", yticks=5).opts(xrotation=90)
gta_graph
```

```python
# Barrie House price from 2019 to 2020 plot
barrie_graph = df_barrie_19_20.hvplot(title = "Barrie House price from 2019 to 2020").opts(xrotation=90)
barrie_graph
```

```python
# Hamilton House price from 2019 to 2020 plot
hamilton_graph = df_hamilton_19_20.hvplot(title = "Hamilton House price from 2019 to 2020").opts(xrotation=90)
hamilton_graph
```

```python
# Read csv contains house prices in Kitchener
df = pd.read_csv("Data/House_price_Kitchener.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)

# Use Kitchener Composite Data
df_kitchener = df[["Composite_Benchmark_SA_Kitchener"]]

# Use Kitchener House Data between 2019 to 2020
df_kitchener_19_20 = df_kitchener["2019-01-01":"2020-09-01"]

df_kitchener_19_20.head()
```

```python
# Kitchener House price from 2019 to 2020 plot
kitchener_graph = df_kitchener_19_20.hvplot(title = "Kitchener House price from 2019 to 2020").opts(xrotation=90)
kitchener_graph
```

```python
# Read csv contains house prices in Niagara
df = pd.read_csv("Data/House_price_Niagara.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)

# Use Niagara Composite Data
df_niagara = df[["Composite_Benchmark_SA_Niagara"]]

# Use Niagara House Data between 2019 to 2020
df_niagara_19_20 = df_niagara["2019-01-01":"2020-09-01"]

df_niagara_19_20.head()
```

```python
# Niagara House price from 2019 to 2020 plot
niagara_graph = df_niagara_19_20.hvplot(title = "Niagara House price from 2019 to 2020").opts(xrotation=90)
niagara_graph
```

```python
# Read csv contains house prices in Peterborough
df = pd.read_csv("Data/House_price_Peterborough.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)

# Use Peterborough Composite Data
df_peterborough = df[["Composite_Benchmark_SA_Peterborough"]]

# Use Peterborough House Data between 2019 to 2020
df_peterborough_19_20 = df_peterborough["2019-01-01":"2020-09-01"]

df_peterborough_19_20.head()
```

```python
# Peterborough House price from 2019 to 2020 plot
peterborough_graph = df_peterborough_19_20.hvplot(title = "Peterborough House price from 2019 to 2020").opts(xrotation=90)
peterborough_graph
```

```python
# Concatenate the house prices for all cities
df_total = pd.concat([df_gta_19_20, 
                      df_barrie_19_20, 
                      df_hamilton_19_20, 
                      df_kitchener_19_20, 
                      df_niagara_19_20, 
                      df_peterborough_19_20], 
                      axis=1, 
                      sort=False)

house_price_ontario_cities = df_total.hvplot(title="House price from 2019 to 2020 in Ontario",
                                             legend="top_left").opts(xrotation=90)

house_price_ontario_cities
```

```python
# Plot violin plot for House prices

houseprice_violin_plot = df_total.hvplot.violin(height=600,
                                             title = "Ontario House Price from 2019-2020").opts(xrotation=90)

houseprice_violin_plot
```

```python
#Plot the Heatmap
fig,ax = plt.subplots(figsize=(8,6))
correlation = df_total.corr()
sns.heatmap(correlation,ax=ax,vmax=1,square=True)
plt.title('Numeric Feature heatmap - Ontario House price')
```

```python
# Read csv contains house prices in GTA
df = pd.read_csv("Data/House_price_GTA.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)
df_gta_HPI = df[["Composite_HPI_GTA"]]

# GTA Housing prices Index for 2019-2020
df_gta_HPI_19_20 = df_gta_HPI["2019-01-01":"2020-09-01"]

df_gta_HPI_19_20.head()
```

```python
# Read csv contains house prices in Barrie
df = pd.read_csv("Data/House_price_Barrie.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)
df_barrie_HPI = df[["Composite_HPI_Barrie"]]

#Barrie housing prices index for 2019-2020
df_barrie_HPI_19_20 = df_barrie_HPI["2019-01-01":"2020-09-01"]

df_barrie_HPI_19_20.head()
```

```python
# Read csv contains house prices in Hamilton
df = pd.read_csv("Data/House_price_Hamilton.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)
df_hamilton_HPI = df[["Composite_HPI_Hamilton"]]

# Hamilton prices index for 2019-2020
df_hamilton_HPI_19_20 = df_hamilton_HPI["2019-01-01":"2020-09-01"]

df_hamilton_HPI_19_20.head()
```

```python
# Read csv contains house prices in Kitchener
df = pd.read_csv("Data/House_price_Kitchener.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)
df_kitchener_HPI = df[["Composite_HPI_Kitchener"]]

# Kitchener prices index for 2019-2020
df_kitchener_HPI_19_20 = df_kitchener_HPI["2019-01-01":"2020-09-01"]

df_kitchener_HPI_19_20.head()
```

```python
# Read csv contains house prices in Niagara
df = pd.read_csv("Data/House_price_Niagara.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)
df_niagara_HPI = df[["Composite_HPI_Niagara"]]

# Niagara prices index for 2019-2020
df_niagara_HPI_19_20 = df_niagara_HPI["2019-01-01":"2020-09-01"]

df_niagara_HPI_19_20.head()
```

```python
# Read csv contains house prices in Peterborough
df = pd.read_csv("Data/House_price_Peterborough.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)

# Drop null values
df.dropna(inplace=True)
df_peterborough_HPI = df[["Composite_HPI_Peterborough"]]

# Peterborough prices index for 2019-2020
df_peterborough_HPI_19_20 = df_peterborough_HPI["2019-01-01":"2020-09-01"]

df_peterborough_HPI_19_20.head()
```

```python
# Concatenate home price indices of all the cities for 2019-2020
df_total_HPI = pd.concat([df_gta_HPI_19_20, 
                          df_barrie_HPI_19_20, 
                          df_hamilton_HPI_19_20, 
                          df_kitchener_HPI_19_20, 
                          df_niagara_HPI_19_20, 
                          df_peterborough_HPI_19_20], 
                          axis=1, 
                          sort=False)

# Violin Plot the Ontario House prices index from 2019-2020
hpi_violin_plot = df_total_HPI.hvplot.violin(height=600,
                                             title = "Ontario Home Price Index from 2019-2020").opts(xrotation=90)

hpi_violin_plot
```

```python
# HPI plot for 2019-2020
hpi_plot = df_total_HPI.hvplot(legend="top_left",title = "Ontario House Price from 2019-2020").opts(xrotation=90)
hpi_plot
```

```python
# Calculate percent increase during Covid
df_during_covid = df_total["2020-03-01":"2020-09-01"]
percent_increase = ((df_during_covid.iloc[len(df_during_covid)-1]/df_during_covid.iloc[0])-1)*100
df_percent_increase = percent_increase.to_frame()

df_percent_increase.columns = ["Percent increase during Covid"]

# Plot Percent increase during covid
df_percent_increase_plot = df_percent_increase.hvplot.bar(title = "Percent Increase - House price in Ontario During COVID",
                                                          height=500).opts(xrotation=90)

df_percent_increase_plot
```
